`````
import pandas as pd

PUBID = 'pubid'
AUTHID = 'authid'
TYPE = 'TIPO_PRODUCAO'
YEAR = 'ANO'
TITLE = 'TITULO'
DOI = 'DOI'

kennis_03 = pd.read_csv('kennis_ids_03.csv')
kennis_04 = pd.read_csv('kennis_ids_04.csv')
kennis_05 = pd.read_csv('kennis_ids_05.csv')
kennis_06 = pd.read_csv('kennis_ids_06.csv')
kennis_07 = pd.read_csv('kennis_ids_07.csv')
kennis_08 = pd.read_csv('kennis_ids_08.csv')
kennis_09 = pd.read_csv('kennis_ids_09.csv')
kennis_10 = pd.read_csv('kennis_ids_10.csv')
`````

`````
def not_single(kennis):
    return lambda publication: len(kennis[kennis[PUBID] == publication.pubid]) > 1

def shared_pubs(kennis):
    datacols = kennis[kennis.apply(not_single(kennis), axis=1)]
    return datacols.sort_values(by=[PUBID, YEAR, TYPE, DOI])
`````

`````
def not_single(kennis):
    def publication_cluster(publication):
        return kennis[kennis[PUBID] == publication.pubid]    

    def authors(cluster):
        return cluster[AUTHID].unique()

    def not_single_filter(publication):
        cluster = publication_cluster(publication)

        return len(cluster) > 1 and len(authors(cluster)) > 1

    return not_single_filter

def shared_pubs(kennis):
    datacols = kennis[kennis.apply(not_single(kennis), axis=1)]
    return datacols.sort_values(by=[PUBID,YEAR, AUTHID, TYPE, DOI])
`````

`````
shared_03 = shared_pubs(kennis_03)
shared_04 = shared_pubs(kennis_04)
shared_05 = shared_pubs(kennis_05)
shared_06 = shared_pubs(kennis_06)
shared_07 = shared_pubs(kennis_07)
shared_08 = shared_pubs(kennis_08)
shared_09 = shared_pubs(kennis_09)
shared_10 = shared_pubs(kennis_10)
`````
